```python
import pandas as pd
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/Migueuu17/alturapse/refs/heads/main/altura-peso.csv'
dados = pd.read_csv(url, sep=',')
dados
```

```python
media_altura = dados['Altura(cm)'].mean(numeric_only=True)
media_altura
```

```python
media_peso = dados['Peso(kg)'].mean()
media_peso
```

```python
dados['Altura em unidade-padrao'] = ((dados['Altura(cm)'] - media_altura)/np.std(dados['Altura(cm)'])).round(2)
```

```python
dados['Peso em unidade-padrao'] = ((dados['Peso(kg)'] - media_peso)/np.std(dados['Peso(kg)'])).round(2)
```

```python
dados['Peso em unidade-padrao x Altura em unidade-padrao'] = (dados['Peso em unidade-padrao']*dados['Altura em unidade-padrao']).round(2)
```

```python
Coef_correlacao = (1/len(dados["Aluno"]))*sum(dados['Peso em unidade-padrao x Altura em unidade-padrao'])
```
